# 3.3.4 Descriptive Statistics and Normality Drill Workspace


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Distribution 1: Gamma (shape, scale, size)

```python
gamma = np.random.gamma(10, 2, 1000)

plt.hist(gamma, bins=20)
plt.axvline(gamma.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(gamma.mean() + gamma.std(), linestyle='dashed', linewidth=2, color='red')
plt.axvline(gamma.mean() - gamma.std(), linestyle='dashed', linewidth=2, color='red')
plt.show()
```

Summary: In order for the mean and standard deviation descriptors to provide useful information about the variable, the data cannot be skewed in any way.  Depending on the distribution's parameters, the distributions that have somewhat useful mean and standard deviation values are negative binomial, and logistic.  In general, unless a variable is normally distributed, mean and standard deviation are not good descriptors.  For the most part, they should be described by their individual parameters instead.

```python
#Two normally distributed variables
var1 = np.random.normal(5, 0.5, 1000)
var2 = np.random.normal(10, 1, 1000)

#added together
var3 = var1 + var2

plt.hist(var3, bins=20)
plt.axvline(var3.mean(), linestyle='solid', linewidth=2, color='red')
plt.axvline(var3.mean() + var3.std(), linestyle='solid', linewidth=2, color='red')
plt.axvline(var3.mean() - var3.std(), linestyle='solid', linewidth=2, color='red')

print('The mean is {}'.format(var3.mean()))
print('The standard deviation is {}'.format(var3.std()))
```

Summary: The mean of var3 is the sum of the two means.  The standard deviation of var3 is not just the sum of the standard deviations, but a mix of the two. After running this simulation a number of times, the higher then number of samples, the more normal the distribution, however with a lower number of samples, the distribution is more skewed.
